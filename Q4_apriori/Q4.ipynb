```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [
    ['A', 'B', 'C', 'D', 'F', 'H'], 
    ['B', 'E', 'F', 'H'],
    ['A', 'C', 'E'],
    ['B', 'C', 'D', 'F', 'H'], 
    ['A', 'B', 'C', 'D', 'E'], 
    ['C','D','F','H'], 
    ['A','C','D','H'],
    ['E','H']
]
```

```python
encoder = TransactionEncoder()
transactions = encoder.fit_transform(dataset)
data = pd.DataFrame(transactions, columns=encoder.columns_) 
data
```

```python
frequent_itemsets = apriori(data, min_support=0.5, use_colnames=True) 
frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric='confidence', min_threshold=0.75)
```

```python
frequent_itemsets = apriori(data, min_support=0.6, use_colnames=True) 
frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load the perfume dataset
df = pd.read_csv('perfume_data.csv')

# Transform the data into a transaction matrix
te = TransactionEncoder()
te_ary = te.fit_transform(df.values)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Use minimum support of 50% and minimum confidence of 75%
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.75)

# Print the frequent itemsets and association rules
print("Minimum support: 50%, Minimum confidence: 75%")
print("Frequent itemsets:")
print(frequent_itemsets)
print("\nAssociation rules:")
print(rules)

# Use minimum support of 60% and minimum confidence of 60%
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Print the frequent itemsets and association rules
print("\nMinimum support: 60%, Minimum confidence: 60%")
print("Frequent itemsets:")
print(frequent_itemsets)
print("\nAssociation rules:")
print(rules)
```
